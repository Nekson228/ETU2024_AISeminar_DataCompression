```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cov = np.array([
    [5, 0], 
    [0, 0.001]
])
mean = np.array([0, 0])
data = np.random.multivariate_normal(mean, cov, size=1000)
data
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica"
})
```

```python
plt.scatter(data[:, 0], data[:, 1])
plt.ylim(-5, 5)

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Original Data')

plt.savefig('../../resources/pca/simple_example.png', transparent=True)
```

```python
B = np.array([1, 0]).reshape(-1, 1)
B
```

```python
transformed = B.T @ data.T
decoded = (B @ transformed).T
decoded
```

```python
plt.scatter(decoded[:, 0], decoded[:, 1], color='orange')
plt.ylim(-5, 5)

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Compressed Data')

plt.savefig('../../resources/pca/simple_example_comp.png', transparent=True)
```
